```python
from glob import glob
import os
import inspect
import requests
# import urllib.request
from urlparse import urlparse
import wget
```

```python
STARTING_DIR = os.getcwd()
```

```python
URLS_START_RANGE = 1
URLS_END_RANGE = 101
```

```python
def fetch_urls(path):
    os.chdir(path)
    urls = [line.strip() for line in open('urls')]
    for url in urls[URLS_START_RANGE:URLS_END_RANGE]:
        u = url_parts = urlparse(url)
        #filename = os.path.basename(url_parts.path)
        hostname = u.hostname #.replace('/', '_')
        if u.port:
            hostname += '_' + str(u.port)
        path = u.path.replace('/', '_').replace('.JPG', '.jpg')
        filename = '%s_%s%s' % (u.scheme, hostname, path)
        #print('filename: %s' % filename)
        extension = os.path.splitext(filename)[1]
        if not extension.lower() == '.jpg':
            print('not downloading ' + url)
            continue
        if os.path.exists(filename):
            #print('already downloaded: ' + url)
            continue
        try:
            r = requests.get(url, timeout=1)
            print('fetched ' + url)
            with open(filename, 'wb') as f:  
                f.write(r.content)
                f.close()
        except:
            print('failed to load ' + url)

    os.chdir(STARTING_DIR)

```

```python
PATH_TO_IMAGE_DIRS = 'produce'
IMAGE_DIRS = glob(os.path.join(STARTING_DIR, PATH_TO_IMAGE_DIRS, '*'))
IMAGE_DIRS.sort()
for dir in IMAGE_DIRS:
    print('\n====================================')
    print(dir)
    print('====================================')
    fetch_urls(dir)
    #break
```
